# Reshaping

Pivot: The pivot function in pandas is used to reshape your DataFrame by converting columns into rows to get a new derived table. Itâ€™s useful when you want to generate a simple cross-tabulation of two (or more) factors. It takes three arguments: index, columns, and values. As a result, a new table is derived which has unique values from index in the index axis and unique values from columns in the columns axis. values fill up the DataFrame.

Melt: The melt function in pandas is used to transform or reshape data. Itâ€™s used to unpivot a DataFrame from wide format to long format. The id_vars represent the columns of the data you do not want to melt, and value_vars represent the columns you do wish to melt into rows. By default, if no value_vars are provided, all columns not set in the id_vars will be melted.

Wide Format: In a wide format, each subjectâ€™s repeated responses will be in a separate column. For example, if you have sales data for apples and bananas on different dates, in wide format, you might have a separate column for apple sales and banana sales, with each row representing a different date.

Long Format: In a long format, each row is a one-time point per subject. So, using the same fruit sales example, in long format, you would have a column for the date, a column for the type of fruit, and a column for the sales, with each row representing one type of fruitâ€™s sales data for a particular date.


pd.stack():pivot the level of an index in a DataFrame

```python
data = {
    'Date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02'],
    'Product': ['Apple', 'Banana', 'Apple', 'Banana'],
    'Sales': [100, 150, 120, 180]
}

#This is a long format
df_example = pd.DataFrame(data)
display(df_example)
```

```python
#This will give you a DataFrame with dates as the index, different fruits as columns,
# and sales as the values. This is a wide format
pivot_df = df_example.pivot(index='Date', columns='Product', values='Sales')
display(pivot_df)
```

```python
#reverts to its original form
melted_df = pivot_df.reset_index().melt(id_vars='Date', var_name='Product', value_name='Sales')
display(melted_df)
```

```python
#Creates a Series with a MultiIndex and a 'Sales' column
stacked_df = melted_df.set_index(['Date', 'Product']).stack().reset_index(name='Sales')
display(stacked_df)
```

In Python, and is used when you want to check if two things are both true. For example, if you have two light switches, 
and checks if both are on.

But when youâ€™re dealing with lists of things (like a column in a spreadsheet),
you need to check each item one by one. Thatâ€™s where & comes in. 
Itâ€™s like having a helper who goes through each row and checks if both conditions are true.

So in this code, & is used because we're checking each row in your DataFrame to see if it meets
both conditions: â€˜Length [m]â€™ is greater than 10 and â€˜Event Descriptionâ€™ is â€˜SOSâ€™. 
If you used and, Python would get confused because itâ€™s not clear what it means for a whole list of things
to be â€˜trueâ€™ or â€˜falseâ€™.

So, in the context of filtering pandas DataFrames based on multiple conditions,
you should use & (and | for OR, ~ for NOT), not and (or or, not)


```python
# Filter rows based on multiple conditions
filtered_data = df[(df['Length [m]'] > 10) & (df['Event Description'] == 'SOS')]
display(filtered_data)
```

```python
# Select rows where length is greater than 20 and specific columns using .loc[]
selected_data = df.loc[df['Length [m]'] > 20, ['Event Description', 'Span Height [m]', 'Overall Pipe Diameter [m]']]
selected_data
```

```python
# Using iloc to select specific rows and columns by integer positions
# Selecting the first and third rows, and the second and third columns
selected_data = df.iloc[[0, 2], [1, 2]]
selected_data
```

can use .iat or .at to find single values
.iat - Access a single value for a row/column pair by integer position.
.at - Access a single value for a row/column label pair.
Note that .at and .iat are faster than .loc and .iloc for accessing a scalar

```python
data = {'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
        'Value': [10, 15, 20, 25, 30, 35]}
df1 = pd.DataFrame(data)
display(df1)
```

```python
display(df1.iat[3, 1])  # Accessing the value at the fourth row (index 3) and the second column (index 1)
```

```python
display(df1.at[4, 'Value'])  # Accessing the value at the row labeled 4 and column labeled 'Value'
```

```python
display(df1['Value'][4]) #can also index the column using [] as earlier
```
