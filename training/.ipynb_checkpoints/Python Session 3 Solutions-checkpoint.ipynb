```python
import pandas as pd
```

```python
data = {
    'Semester': ['Semester1', 'Semester1', 'Semester2', 'Semester2', 'Semester 1', 'Semester 2'],
    'Subject': ['Math', 'Science', 'Math', 'Science', 'English', 'English'],
    'Score': [85, 90, 88, 92, 71, 75]
}

df = pd.DataFrame(data)
display(df)

#pivot this data to see the scores in each subject across different semesters
#revert the pivoted DataFrame back to its original form
#create a dataframe with a MultiIndex ('Semester', 'Subject') and a 'Score' column
#sort by subject
#sort by semester then score
#sort by descending index
#display only math scores using iloc
#display only semester 1 marks above 80 using square brackets
#reset index based on this
```

```python
#answers for first one
pivot_df = df.pivot(index='Semester', columns='Subject', values='Score')
display(pivot_df)
melted_df = pivot_df.reset_index().melt(id_vars='Semester', var_name='Subject', value_name='Score')
display(melted_df)
# stacked_df = melted_df.set_index(['Semester', 'Subject']).stack().reset_index(name='Score')
# display(stacked_df)
```

```python
selected_data = df.iloc[:3, 1:3]
display(selected_data)
```

```python
selected_data = df.loc[df['Score'] > 80, ['Subject', 'Score']]
display(selected_data)
```

```python
df.iat[2, 2] = 95
display(df)
```

```python
value = df.at[3, 'Score']
print(value)
```

```python
sorted_data = df.sort_values(by='Score', ascending=False)
display(sorted_data)
```

```python
sorted_data = df.sort_values(by=['Semester', 'Subject'], ascending=[True, False])
display(sorted_data)
```

```python
sorted_data = df.sort_index(ascending=False)
display(sorted_data)
```
