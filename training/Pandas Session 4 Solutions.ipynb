```python
import pandas as pd
```

```python
df3 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df4 = pd.DataFrame({'C': [5, 6], 'D': [7, 8]})

display(df3)
display(df4)

# Concatenate df3 and df4 along columns
result_concat = pd.concat([df3, df4], axis=1)

# Display the result
display(result_concat)
```

```python
import pandas as pd

df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [3, 4, 5], 'Age': [25, 30, 22]})

# Combine df1 and df2 using merge
result = df1.merge(df2, on='ID', how='outer')

# Display the result
display(result)
```

```python
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [4, 5, 6], 'Name': ['David', 'Eve', 'Frank']})

display(df1)
display(df2)

# Concatenate df1 and df2 along rows
result_concat = pd.concat([df1, df2], axis=0)

# Display the result
display(result_concat)
```

```python
import pandas as pd

df1 = pd.DataFrame({'Values': [10, 15, 20, 25, 30]})
# Create a new column 'Double_Values' with double of 'Values'
df1['Double_Values'] = df1['Values'].apply(lambda x: 2 * x)

# Display the updated DataFrame
display(df1)
```

```python
df2 = pd.DataFrame({'Scores': [90, 60, 40, 75, 55]})


# Define the categorize_scores function
def categorize_scores(score):
    return 'High' if score >= 50 else 'Low'

# Apply the function to create 'Score_Category' column
df2['Score_Category'] = df2['Scores'].apply(categorize_scores)

# Display the updated DataFrame
display(df2)
```

```python
sales_data = pd.DataFrame({
    'Product': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Sales': [100, 150, 120, 200, 180, 220],
})

grouped_sales = sales_data.groupby('Product').mean()
display(grouped_sales)
```
