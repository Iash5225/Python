# Dropping duplicates

```python
#Drop duplicates 
df_dropped_duplicates = df_excel.drop_duplicates(subset=['Spheroidal KP [km]','Event Description'], keep='first')
display(df_dropped_duplicates)

#subset=['Spheroidal KP [km]']: The subset parameter is used to specify the column(s) for identifying duplicates. 
# In this case, it’s set to ['Spheroidal KP [km]'], which means the function will look for duplicates in this column.

# keep='first': The keep parameter is used to determine which duplicate(s) to keep.
# 'first' means the function will keep the first occurrence of a duplicate and remove the others.
# If it was set to 'last', the function would keep the last occurrence and remove the others. 
# If it was set to False, the function would remove all duplicates.


```
